```python
XboxSales = pd.read_csv('XboxOne_GameSales.csv', sep=',', encoding='cp1252')
```

```python
PS4Sales = pd.read_csv('PS4_GamesSales.csv', sep=',', encoding='cp1252')
```

```python
XboxSales = XboxSales.drop(["Japan", 'Europe', 'Rest of World', 'North America'], axis=1)
```

```python
PS4Sales = PS4Sales.drop(['Europe', 'Rest of World', 'Japan', 'North America'], axis=1)
```

```python
XboxSales.head() , PS4Sales.head()
```

```python
XboxSales = XboxSales.dropna(subset=['Publisher', 'Year', 'Global', 'Game', 'Genre'])
```

```python
PS4Sales = PS4Sales.dropna(subset=['Publisher', 'Year', 'Global', 'Game', 'Genre'])
```

```python
len(XboxSales['Global']), len(PS4Sales['Global'])
```

```python
len(XboxSales[XboxSales['Global']== 0.00]), len(PS4Sales[PS4Sales['Global']== 0.00]) #To drop or not to drop....
```

```python
XboxSales = XboxSales[XboxSales['Global'] != 0.0]
PS4Sales = PS4Sales[PS4Sales['Global'] != 0.0]
```

```python
XboxSales.hist('Global'), PS4Sales.hist('Global')
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
XboxSales['Genre_encoded'] = label_encoder.fit_transform(XboxSales['Genre'])
PS4Sales['Genre_encoded'] = label_encoder.fit_transform(PS4Sales['Genre'])
```

```python
XboxSales.head()
```

```python
colors = plt.cm.viridis(XboxSales['Genre_encoded'] / max(XboxSales['Genre_encoded']))

fig, ax = plt.subplots()
bars = ax.bar(XboxSales['Genre_encoded'], XboxSales['Global'], color=colors)

ax.set_xticks(range(16))

plt.xlabel('Genre')
plt.ylabel('Global')
plt.title('Genres Sales')

plt.show()
```

```python
colors_xbox = plt.cm.viridis(XboxSales['Genre_encoded'] / max(XboxSales['Genre_encoded']))
colors_ps4 = plt.cm.viridis(PS4Sales['Genre_encoded'] / max(PS4Sales['Genre_encoded']))

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))


bars_xbox = axes[0].bar(XboxSales['Genre_encoded'], XboxSales['Global'], color=colors_xbox)
axes[0].set_title('Xbox Sales')
axes[0].set_xticks(range(16))


bars_ps4 = axes[1].bar(PS4Sales['Genre_encoded'], PS4Sales['Global'], color=colors_ps4)
axes[1].set_title('PS4 Sales')
axes[1].set_xticks(range(17))

for ax in axes:
    ax.set_xlabel('Genre Encoded')
    ax.set_ylabel('Global')


fig.suptitle('Genre Sales')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

for genre, color in zip(XboxSales['Genre'].unique(), plt.cm.tab20.colors):
    genre_data = XboxSales[XboxSales['Genre'] == genre]
    axes[0].bar(genre_data['Genre_encoded'], genre_data['Global'], color=color, label=genre, alpha=0.7)
    axes[0].legend(loc='upper left', bbox_to_anchor=(1, 1))

axes[0].set_xlabel('Genre Encoded')
axes[0].set_ylabel('Global')
axes[0].set_title('Xbox Sales')
axes[0].legend()

for genre, color in zip(PS4Sales['Genre'].unique(), plt.cm.tab20.colors):
    genre_data = PS4Sales[PS4Sales['Genre'] == genre]
    axes[1].bar(genre_data['Genre_encoded'], genre_data['Global'], color=color, label=genre, alpha=0.7)
    axes[1].legend(loc='upper left', bbox_to_anchor=(1, 1))

axes[1].set_xlabel('Genre Encoded')
axes[1].set_ylabel('Global')
axes[1].set_title('PS4 Sales')
axes[1].legend()
axes[1].set_xticks(range(17))
axes[0].set_xticks(range(16))
axes[1].set_yticks(range(20))
axes[0].set_yticks(range(20))

plt.tight_layout()


plt.show()
```

```python
bar_width = 0.35
bar_positions_xbox = XboxSales['Year'] - bar_width / 2
bar_positions_ps4 = PS4Sales['Year'] + bar_width / 2

plt.bar(bar_positions_xbox, XboxSales['Global'], width=bar_width, label='Xbox Sales')
plt.bar(bar_positions_ps4, PS4Sales['Global'], width=bar_width, label='PS4 Sales')

plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Game Sales Over Years')
plt.xticks(XboxSales['Year'])

plt.legend()

plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

XboxSales['Year'].value_counts().plot(kind='pie', ax=axes[0])

PS4Sales['Year'].value_counts().plot(kind='pie', ax=axes[1])

axes[0].set_ylabel('')
axes[1].set_ylabel('')
fig.suptitle('Test')
plt.tight_layout()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(PS4Sales['Genre_encoded'], PS4Sales['Year'], PS4Sales['Global'])
ax.set_xlabel('Genre_encoded')
ax.set_ylabel('Year')
ax.set_zlabel('Global')
ax.set_title('PS4Sales')
plt.show()

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(XboxSales['Genre_encoded'], XboxSales['Year'], XboxSales['Global'])
ax.set_xlabel('Genre_encoded')
ax.set_ylabel('Year')
ax.set_zlabel('Global')
ax.set_title('XboxSales')

axes[0].set_xticks(range(17))
axes[1].set_xticks(range(16))

plt.show()
```
