```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(12,6))

GameSales['Year'].plot.hist(ax=axes[0], bins=50, alpha=0.5, color='red')
axes[0].set_title('Year')

GameSales['Europe'].plot.hist(ax=axes[2], bins=50, alpha=.5, color='red')
axes[2].set_title('Europe')

GameSales['Global'].plot.hist(ax=axes[1], bins=50, alpha=.5, color='blue')
axes[1].set_title('Global')


```

```python
np.mean(GameSales['Global'])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

GameSales['Publisher'].value_counts().plot(kind='pie', ax=axes[0])

GameSales['Genre'].value_counts().plot(kind='pie', ax=axes[1])

axes[0].set_ylabel('')
axes[1].set_ylabel('')
fig.suptitle('Test')
plt.tight_layout()
```
